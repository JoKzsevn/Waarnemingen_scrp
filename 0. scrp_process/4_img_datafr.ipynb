```python
import pandas as pd
import os

# Map waar de afbeeldingen zijn opgeslagen
IMAGE_FOLDER = 'downloaded_images'

# Lijst om bestandsnamen en paden op te slaan
image_data = []

# Loop door alle mappen en submappen
for root, dirs, files in os.walk(IMAGE_FOLDER):
    for filename in files:
        if filename.lower().endswith(".jpg"):  # Controleer of het een afbeelding is (case insensitive)
            file_path = os.path.join(root, filename)
            
            # Label is de bestandsnaam zonder .jpg extensie
            label = os.path.splitext(filename)[0]
            
            image_data.append([label, file_path])

# Maak een DataFrame
df_images = pd.DataFrame(image_data, columns=['Label', 'Foto (loc)'])

# Toon het DataFrame
print(f"Aantal gevonden afbeeldingen: {len(df_images)}")
df_images
```

```python
# Optioneel: Opslaan als CSV
df_images.to_csv('image_data.csv', index=False)
```
